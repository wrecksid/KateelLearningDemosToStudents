# Bank Customer Data Generation

This notebook generates synthetic bank customer data using the Faker library. The data follows realistic patterns found in banking datasets and can be used for customer segmentation analysis.

## Features Generated:
- Customer demographics (age, gender, location)
- Financial information (income, account balance, credit score)
- Banking behavior (products owned, transaction frequency, tenure)
- Engagement metrics (digital engagement score)
- Customer lifetime value and churn indicators

## Data Generation Parameters

Adjust these parameters to customize your synthetic dataset:

```python
# Configuration parameters
NUM_RECORDS = 10000        # Number of customer records to generate
MIN_AGE = 18              # Minimum customer age
MAX_AGE = 80              # Maximum customer age
MIN_INCOME = 150000       # Minimum annual income (INR)
MAX_INCOME = 5000000      # Maximum annual income (INR)
CHURN_RATE = 0.15         # Expected churn rate (15%)
LOCALE = 'en_IN'          # Locale for Indian data

print(f"Configuration:")
print(f"- Records to generate: {NUM_RECORDS:,}")
print(f"- Age range: {MIN_AGE}-{MAX_AGE} years")
print(f"- Income range: ₹{MIN_INCOME:,} - ₹{MAX_INCOME:,}")
print(f"- Expected churn rate: {CHURN_RATE:.1%}")
print(f"- Locale: {LOCALE}")
```

## Generate Synthetic Data

```python
# Initialize the data generator
generator = BankDataGenerator(locale=LOCALE, seed=42)

# Generate customer data
print("🔄 Generating synthetic bank customer data...")
df = generator.generate_customer_data(
    num_records=NUM_RECORDS,
    min_age=MIN_AGE,
    max_age=MAX_AGE,
    min_income=MIN_INCOME,
    max_income=MAX_INCOME,
    churn_rate=CHURN_RATE
)

print(f"✅ Generated {len(df)} customer records!")
print(f"Dataset shape: {df.shape}")
```

## Data Quality Checks

```python
# Check for missing values
print("🔍 Missing Values Check:")
missing_values = df.isnull().sum()
if missing_values.sum() == 0:
    print("✅ No missing values found!")
else:
    print(missing_values[missing_values > 0])

# Check for duplicates
duplicates = df.duplicated().sum()
print(f"\n🔍 Duplicate Records: {duplicates}")
if duplicates == 0:
    print("✅ No duplicate records found!")

# Data type validation
print("\n🔍 Data Types:")
print(df.dtypes)
```

## Data Visualizations

```python
# Create comprehensive visualizations
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('Bank Customer Data Overview', fontsize=16, fontweight='bold')

# Age distribution
axes[0,0].hist(df['age'], bins=30, alpha=0.7, color='skyblue', edgecolor='black')
axes[0,0].set_title('Age Distribution')
axes[0,0].set_xlabel('Age')
axes[0,0].set_ylabel('Frequency')

# Income distribution
axes[0,1].hist(df['annual_income'], bins=30, alpha=0.7, color='lightgreen', edgecolor='black')
axes[0,1].set_title('Annual Income Distribution')
axes[0,1].set_xlabel('Annual Income (₹)')
axes[0,1].set_ylabel('Frequency')

# Account balance distribution
axes[0,2].hist(df['account_balance'], bins=30, alpha=0.7, color='orange', edgecolor='black')
axes[0,2].set_title('Account Balance Distribution')
axes[0,2].set_xlabel('Account Balance (₹)')
axes[0,2].set_ylabel('Frequency')

# Credit score distribution
axes[1,0].hist(df['credit_score'], bins=30, alpha=0.7, color='pink', edgecolor='black')
axes[1,0].set_title('Credit Score Distribution')
axes[1,0].set_xlabel('Credit Score')
axes[1,0].set_ylabel('Frequency')

# Products distribution
product_counts = df['num_products'].value_counts().sort_index()
axes[1,1].bar(product_counts.index, product_counts.values, alpha=0.7, color='purple')
axes[1,1].set_title('Number of Products Distribution')
axes[1,1].set_xlabel('Number of Products')
axes[1,1].set_ylabel('Number of Customers')

# Churn distribution
churn_counts = df['is_churned'].value_counts()
axes[1,2].pie(churn_counts.values, labels=['Active', 'Churned'], autopct='%1.1f%%', 
              colors=['lightblue', 'salmon'])
axes[1,2].set_title('Customer Churn Distribution')

plt.tight_layout()
plt.show()

print("✅ Visualizations created successfully!")
```

```python
# Additional correlation analysis
plt.figure(figsize=(12, 10))

# Select numerical columns for correlation
numerical_cols = ['age', 'annual_income', 'account_balance', 'credit_score', 
                  'num_products', 'tenure_years', 'monthly_transactions', 
                  'digital_engagement_score', 'customer_lifetime_value']

correlation_matrix = df[numerical_cols].corr()

# Create heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, linewidths=0.5)
plt.title('Feature Correlation Matrix', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print("✅ Correlation analysis completed!")
```
