## Data Exploration and Validation

```python
# Basic statistics
print("Dataset Info:")
print(data.info())
print("\nSummary Statistics:")
data.describe()
```

```python
# Visualize data distributions
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# Income distribution
axes[0, 0].hist(data['annual_income'], bins=50, alpha=0.7)
axes[0, 0].set_title('Annual Income Distribution')
axes[0, 0].set_xlabel('Income ($)')

# Age distribution
axes[0, 1].hist(data['age'], bins=30, alpha=0.7)
axes[0, 1].set_title('Age Distribution')
axes[0, 1].set_xlabel('Age')

# Account balance distribution
axes[0, 2].hist(data['account_balance'], bins=50, alpha=0.7)
axes[0, 2].set_title('Account Balance Distribution')
axes[0, 2].set_xlabel('Balance ($)')

# Credit score distribution
axes[1, 0].hist(data['credit_score'], bins=30, alpha=0.7)
axes[1, 0].set_title('Credit Score Distribution')
axes[1, 0].set_xlabel('Credit Score')

# Transaction frequency
axes[1, 1].hist(data['monthly_transactions'], bins=30, alpha=0.7)
axes[1, 1].set_title('Monthly Transactions Distribution')
axes[1, 1].set_xlabel('Transactions per Month')

# Number of products
axes[1, 2].hist(data['num_products'], bins=8, alpha=0.7)
axes[1, 2].set_title('Number of Products Distribution')
axes[1, 2].set_xlabel('Number of Products')

plt.tight_layout()
plt.show()
```

```python
# Correlation analysis
plt.figure(figsize=(12, 10))
correlation_features = ['age', 'annual_income', 'account_balance', 'credit_score', 
                       'monthly_transactions', 'num_products', 'tenure_years']
correlation_matrix = data[correlation_features].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix')
plt.show()
```

## Parameter Tuning Example

Students can experiment with different parameters to see how they affect the data generation.

```python
# Example: Modify parameters for different scenarios

# Scenario 1: High-income customer base
generator_high_income = BankingDataGenerator(seed=42)
generator_high_income.update_parameters({
    'income_range': (80000, 500000),
    'account_balance_range': (10000, 500000)
})

data_high_income = generator_high_income.generate_customer_data(1000)

# Compare distributions
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

axes[0].hist(data['annual_income'], bins=30, alpha=0.7, label='Original')
axes[0].hist(data_high_income['annual_income'], bins=30, alpha=0.7, label='High Income')
axes[0].set_title('Income Distribution Comparison')
axes[0].set_xlabel('Annual Income ($)')
axes[0].legend()

axes[1].hist(data['account_balance'], bins=30, alpha=0.7, label='Original')
axes[1].hist(data_high_income['account_balance'], bins=30, alpha=0.7, label='High Income')
axes[1].set_title('Account Balance Distribution Comparison')
axes[1].set_xlabel('Account Balance ($)')
axes[1].legend()

plt.tight_layout()
plt.show()
```
